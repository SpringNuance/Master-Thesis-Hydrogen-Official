The non-smooth crack phase field can be approximated by the exponential function

$\phi(x)=\mathrm{e}^{-\frac{|x|}{\ell}}$


$\Gamma_{\ell}(\phi)=\int_{\Omega}\left(\dfrac{1}{2 \ell} \phi^2+\dfrac{\ell}{2}|\nabla \phi|^2\right) \mathrm{d} V=\int_{\Omega} \gamma_{\ell}(\phi, \nabla \phi) \mathrm{d} V$

$\int_{\Gamma} G_c(\theta) \mathrm{d} S \approx \int_{\Omega} G_c(\theta)\left(\dfrac{1}{2 \ell} \phi^2+\dfrac{\ell}{2}|\nabla \phi|^2\right) \mathrm{d} V$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1001)
length_scale = [5, 2, 1, 0.5, 0.1, 0.01]

plt.figure(figsize=(8, 5))

for l in length_scale:
    phi_x = np.exp(-np.abs(x)/l)
    # set phi_x = 1 where x = 0
    phi_x[np.where(x == 0)] = 1
    plt.plot(x, phi_x, label=rf'$\ell$ = {l}')

plt.legend()
plt.xlabel('x')
plt.ylabel(r'$\phi(x)$')
plt.title('Smooth diffusive crack topology function')
```
