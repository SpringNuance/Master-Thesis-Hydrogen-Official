```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Cbar_L_CPE8RT_data = np.loadtxt('result_CPE8RT/export_data/Cbar_L.txt', skiprows=2)
Cbar_L_CPE8HT_data = np.loadtxt('result_CPE8HT/export_data/Cbar_L.txt', skiprows=2)

columns=['X', 'Cbar_L_Point_A', 'Cbar_L_Point_A', 'Cbar_L_Point_A']

df_CPE8RT = pd.DataFrame(data = Cbar_L_CPE8RT_data, columns=columns)
df_CPE8HT = pd.DataFrame(data = Cbar_L_CPE8HT_data, columns=columns)

time = df_CPE8RT.iloc[:, 0].to_numpy()
Cbar_L_point_A_CPE8RT = df_CPE8RT.iloc[:, 1].to_numpy()   
Cbar_L_point_B_CPE8RT = df_CPE8RT.iloc[:, 2].to_numpy()
Cbar_L_point_C_CPE8RT = df_CPE8RT.iloc[:, 3].to_numpy()

Cbar_L_point_A_CPE8HT = df_CPE8HT.iloc[:, 1].to_numpy()
Cbar_L_point_B_CPE8HT = df_CPE8HT.iloc[:, 2].to_numpy()
Cbar_L_point_C_CPE8HT = df_CPE8HT.iloc[:, 3].to_numpy()
```

```python
# Create subplot of 2 x 1

fig, ax = plt.subplots(2, 1, figsize=(15, 10))

# Make fontsize bigger
plt.rcParams.update({'font.size': 14})

# make tick size bigger
ax[0].tick_params(axis='both', which='major', labelsize=14)



fig.suptitle(r'Concentration of hydrogen at lattice sites evolution')
ax[0].plot(time, Cbar_L_point_A_CPE8RT, label='Point A (CPE8RT)', color='blue')
ax[0].plot(time, Cbar_L_point_B_CPE8RT, label='Point B (CPE8RT)', color='red')
ax[0].plot(time, Cbar_L_point_C_CPE8RT, label='Point C (CPE8RT)', color='green')

# We need to annotate the last value Cbar_L for each point at the end of the curve, and offset it below a little
ax[0].annotate(f'{round(Cbar_L_point_A_CPE8RT[-1], 6)}', (time[-1], Cbar_L_point_A_CPE8RT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')
ax[0].annotate(f'{round(Cbar_L_point_B_CPE8RT[-1], 6)}', (time[-1], Cbar_L_point_B_CPE8RT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')
ax[0].annotate(f'{round(Cbar_L_point_C_CPE8RT[-1], 6)}', (time[-1], Cbar_L_point_C_CPE8RT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')

ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel(r'$\bar{C}_L$ [mol/m]')

ax[0].legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)

ax[0].grid(True)

ax[1].plot(time, Cbar_L_point_A_CPE8HT, label='Point A (CPE8HT)', color='blue')
ax[1].plot(time, Cbar_L_point_B_CPE8HT, label='Point B (CPE8HT)', color='red')
ax[1].plot(time, Cbar_L_point_C_CPE8HT, label='Point C (CPE8HT)', color='green')

# We need to annotate the last value Cbar_L for each point at the end of the curve, and offset it below a little
ax[1].annotate(f'{round(Cbar_L_point_A_CPE8HT[-1], 6)}', (time[-1], Cbar_L_point_A_CPE8HT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')
ax[1].annotate(f'{round(Cbar_L_point_B_CPE8HT[-1], 6)}', (time[-1], Cbar_L_point_B_CPE8HT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')
ax[1].annotate(f'{round(Cbar_L_point_C_CPE8HT[-1], 6)}', (time[-1], Cbar_L_point_C_CPE8HT[-1]), textcoords="offset points", xytext=(-50,15), ha='center')

ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel(r'$\bar{C}_L$ [mol/m]')

ax[1].legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)

ax[1].grid(True)
```
