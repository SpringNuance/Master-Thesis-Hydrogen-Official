Oriani’s theory is based on considering a lattice consisting of
two kinds of sites for occupancy by hydrogen. He postulates that
the vast majority of sites are the ordinary or ‘‘normal interstitial”
sites. The minor dfraction of sites called ‘‘trapping” sites provides
an energetically favourable environment for occupancy by the
hydrogen atoms. 

Total concentration of hydrogen. The unit (1/m^3) is also called ppm, and also atoms per cubic unit, usually cubic meter

$C [1/m^3] = C_1 [1/m^3] + C_T [unit: 1/m^3]$

---

$C_1$: number of H atoms per unit volume in the lattice 

$C_1 [1/m^3] = \beta [\text{dimless}] \times \theta_L [\text{dimless}] \times N_L [1/m^3]$

where 
- $\beta [\text{dimless}]$: Number of interstitial sites per solvent atom
- $\theta_L [\text{dimless}]$ in range [0,1] is dfraction of lattice sites occupied by H atoms
- $N_L$ [mol/m^3] (const) = $\dfrac{\rho [mol/m^3] N_A [mol^{-1}]}{A_r [\text{dimless}]}$
- or number of atoms of solvent per unit volume = molar density of lattice element * avogrado / relative atomic mass of the lattice element

Relative atomic mass is unitless it is because it is just the relation between the mass of any atom with respect to the mass of c 12 atom.

Avogradro constant = 6.02214 x 10^23 $mol^{-1}$

---

$C_T [1/m^3] = \alpha [\text{dimless}] \times \theta_T [\text{dimless}] \times N_T [1/m^3]$

where 

- $\alpha [\text{dimless}]$: Number of trapping sites
- $\theta_T [\text{dimless}]$ in range [0,1] is dfraction of trapping sites occupied by H atoms
- $N_T [1/m^3]$ is number of atomic trapping sites along the total dislocation line length per unit volume. It is proportional to the dislocation density ($\rho_d$), which is function of accumulated plastic strain ($\epsilon_p$), where 
- $a [m]$ is the lattice constant and the prefactor is 1/b, where b is the magnitude of the a/2<110> Burgers vector in fcc which is a reasonable approximation for the atomic spacing along a dislocation line. 

$N_T(\epsilon_p) [1/m^3] = \dfrac{\sqrt{2}}{a [m]} \times \rho_d(\epsilon_p) [1/m^2]$

---
It is more convenient when analysing H diffusion in a body to express the concentration in units of $mol/m^3$

$\bar{C}_L [mol/m^3] = C_1 [1/m^3] / N_A [mol^{-1}]$

$\bar{C} [mol/m^3] = C_T [1/m^3] / N_A [mol^{-1}]$

### Mass conservation

Mass conservation states that the rate of change of the total hydrogen concentration in a volume $ \Omega $ is equal to the flux $\mathbf{\bar{J}}$ $(mol \cdot m^{-2} s^{-1})$ through the surface $ \partial\Omega $.

$$\dfrac{\partial}{\partial t} \int_{\Omega} (\bar{C}_L + \bar{C}) d\Omega + \int_{\partial\Omega} \mathbf{\bar{J}} \cdot \mathbf{n} dS = 0 \quad \text{Eq. 7}$$

where $\mathbf{n}$ is the outward normal to the surface $ \partial\Omega $.

The driving force for diffusion is the chemical potential gradient, the flux $\mathbf{\bar{J}}$ can be expressed as follows:


$$\mathbf{\bar{J}} = - \dfrac{D_L C_1}{RT} \nabla \mu \quad \text{Eq. 8}$$  

where $D_L$ is the diffusion coefficient for hydrogen. Here we ignore pipe diffusion along the dislocation line on grounds that the dislocation spacing is much larger than the core radius and that fast’ pipe diffusion may be slower than diffusion through the lattice. The chemical potential, $ \mu $, is defined as follows:

$$\mu = \mu_0 + RT \ln \bar{C}_L + \mu_\sigma \quad \text{Eq. 11}$$ 

$\mu_0$ represents the chemical potential at standard condition, $\mu_\sigma = -\sigma_H V_H$ where $\sigma_H$ is the hydrostatic stress and $V_H$ the partial molar volume of hydrogen. Substituting (9) into (8) we obtain:

$J = \dfrac{D_L V_H \bar{C}_L}{RT} \nabla \sigma_H - D_L \nabla \bar{C}_L \quad \text{Eq. 10}$

Substituting (10) into (7) then gives:

$$\dfrac{\partial}{\partial t} \int_{\Omega} (\bar{C}_L + \bar{C}) d\Omega + \int_{\partial\Omega} \left( \dfrac{D_L V_H \bar{C}_L}{RT} \nabla \sigma_H - D_L \nabla \bar{C}_L \right) \cdot \mathbf{n} dS = 0 \quad \text{Eq. 11}$$

Applying the divergence theorem, we obtain:

$$\dfrac{\partial \bar{C}_L}{\partial t} + \dfrac{\partial \bar{C}}{\partial t} + \nabla \cdot \left( \dfrac{D_L V_H \bar{C}_L}{RT} \nabla \sigma_H \right) - \nabla \cdot (D_L \nabla \bar{C}_L) = 0 \quad \text{Eq. 12}$$ 

---


### The unknowns $\bar{C}_L$ and $\bar{C}$ 

In Eq. (12) there are two unknowns: $C_1$ and $C_T$. However following Oriani’s theory the H concentration in the lattice was assumed to be in equilibrium with the concentration in the traps. This means that once we know the concentration in the lattice we can calculate the concentration in the traps.

$$\bar{C} = \dfrac{N_T \theta_T}{N_A} = \dfrac{N_T}{N_A} \dfrac{1}{1+\dfrac{(1-\theta_T)}{\theta_T}} \quad \text{Eq. 13}$$

Oriani’s theory states that for $\theta_L << 1$:

$$\dfrac{\theta_T}{(1 - \theta_T)} = K \theta_L \quad \text{Eq. 14} $$ 

where

$$K = \exp \left( \dfrac{-W_B}{RT} \right) = \text{const.}  \quad \text{Eq. 15} $$ 

and $ W_B < 0 $ represents the binding energy for the traps. Substituting (14) into (13) we can express the concentration in the traps as a function of the density of trapping sites and the concentration in the lattice:

$$ \bar{C}(C_1, \bar{N}_T) = \bar{N}_T (\epsilon_p) \dfrac{K\theta_L}{1 + K\theta_L} \dfrac{\beta \bar{N}_L}{\beta \bar{N}_L} = \bar{N}_T(\epsilon_p) \dfrac{K \bar{C}_L}{\beta \bar{N}_L + K \bar{C}_L} \text{Eq. 16} $$

where

$$ \bar{N}_L = \dfrac{N_L}{N_A} = \text{const.} \quad \quad \bar{N}_T (\epsilon_p) = \dfrac{N_T (\epsilon_p)}{N_A} \text{Eq. 17}$$

where $\epsilon_p$ is a scalar measure of the plastic strain and $\bar{N}_T(\epsilon_p)$ is a function of the plastic strain through (4) and the constitutive relationship that relates dislocation density to $\epsilon_p$. The time derivative of the trapping site concentration, $\dfrac{\partial \bar{C}}{\partial t}$ in (12), can now be expressed as follows:

$$\dfrac{\partial \bar{C}}{\partial t} = \dfrac{\partial \bar{C}}{\partial \bar{C}_L} \dfrac{\partial \bar{C}_L}{\partial t} + \dfrac{\partial \bar{C}}{\partial \bar{N}_T} \dfrac{d \bar{N}_T}{d\epsilon^p} \dfrac{d\epsilon^p}{dt} \quad \text{Eq. 18}$$

Substituting (18) in (12) we obtain:

$$\dfrac{\partial \bar{C}_L}{\partial t} \left( 1 + \dfrac{\partial \bar{C}}{\partial \bar{C}_L} \right) - \nabla \cdot (D_L \nabla \bar{C}_L) + \nabla \cdot \left( \dfrac{D_L \bar{C}_L V_H}{RT} \nabla \sigma_H \right) + \dfrac{\partial \bar{C}}{\partial \bar{N}_T} \dfrac{d \bar{N}_T}{d\epsilon^p} \dfrac{d\epsilon^p}{dt} = 0 \quad \text{Eq. 19}$$

From (16) we obtain:

$$\dfrac{\partial \bar{C}}{\partial \bar{C}_L} = \dfrac{\beta \bar{N}_T K \bar{N}_L}{(\beta \bar{N}_L + K \bar{C}_L)^2} \quad \text{Eq. 20}$$

$$\dfrac{\partial \bar{C}}{\partial N_T} = \dfrac{K \bar{C}_L}{\beta \bar{N}_L + K \bar{C}_L} \quad \text{Eq. 21}$$

### Numerical results: pure diffusion

We present here the solution of a simple 1D diffusion problem, Fig. 1, which we can compare with the analytical solution. We consider that all H resides in lattice sites, i.e there are no traps and $\bar{C} = \bar{C}_L$. Fig. 1 shows a bar of length $l = 50 mm$. $\bar{C}_L = 0$ throughout the bar and at time $t = 0$ the concentration at one end is increased to $100 mol/m^3$. Fig. 1 shows the full transient solution. The governing equation for this problem is well known as well as the analytic solution [20]:

$$\dfrac{\partial \bar{C}_L^2}{\partial x^2} = \dfrac{1}{D_L} \dfrac{\partial \bar{C}_L}{\partial t} \quad \text{Eq. 26}$$

The analytic solution reads:

$$\bar{C}_L(x, t) = \bar{C}_0 + \dfrac{x}{l} (\bar{C}_l - \bar{C}_0) + \sum_{n=1}^{\infty} c_n exp\left(-\dfrac{n^2 \pi^2 D_L t}{l^2}\right) \sin\left(\dfrac{n \pi x}{l}\right) \quad \text{Eq. 27}$$

and

$$ c_n = \dfrac{2}{l} \int_{0}^{l} f(x) \sin \left( \dfrac{n\pi x}{l} \right) dx + \dfrac{2}{n\pi} \left[ -1^n (\bar{C}_l - \bar{C}_0) \right ] \quad \text{Eq. 28}$$ 


Eq. (26) is essentially a simplified version of (19) in which the term $\dfrac{\partial \bar{C}}{\partial \bar{C}_L} = 0$ and the last two terms of (19) are zero. Eq. (26) is implemented in UMATHT by simply coding the following terms:

$$\dfrac{\partial \bar{C}}{\partial \bar{C}_L} = 1 \quad \text{Eq. 29}$$

$$\dfrac{\partial \mathbf{\bar{J}}_m}{\partial (\nabla \bar{C}_L)} = -D_L \mathbf{I} \quad \text{Eq. 30}$$

Fig. 1 shows the transient hydrogen concentration distribution along the bar for both the analytical and computational solution. The value of diffusivity is $D_L = 3.8 \times 10^{-11} m^2/s$ and the computations were conducted using quadratic continuum coupled temperature-displacements hybrid element i.e. CPE8HT. The characteristic time for diffusion (i.e. time necessary for hydrogen to diffuse along the bar) can be determined by using the well known relation $t = \dfrac{l^2}{D_L} = 6.5 \times 10^7 s$. Here we consider a fixed time step of $dt = 1 \times 10^4 s$ and a total time of $t = 1 \times 10^6 s$. The Abaqus results almost perfectly agree with the analytic solution of (27).

The equation 27 is a PDE describing governing diffusion/heat transfer in one-dimensional heat equation. 

The standard form of the one-dimensional heat or diffusion equation is given by:

$$
\frac{\partial C(x,t)}{\partial t} = D \frac{\partial^2 C(x,t)}{\partial x^2} \quad \text{, where} 
$$



- $ C(x,t) $ is the concentration or temperature at position $ x $ and time $ t $,
- $ D $ is the Hydrogen diffusivity / thermal conductivity (heat transfer version) of the lattice
- $ \frac{\partial}{\partial t} $ and $ \frac{\partial^2}{\partial x^2} $ are the first and second partial derivatives with respect to time and space, respectively.

Equation 27 is obtained by separating variables and applying boundary conditions of $\bar{C}_0(t = 0)$ and $\bar{C}_l(t = 0)$, which is initial conditions for concentration/temperature distribution at begin of the process. The Fourier series expansion satisfies these conditions.



$
\text{The exponential term in the equation}: \quad \quad  \exp\left(-\frac{n^2\pi^2Dt}{l^2}\right)
$

is the time-dependent, which shows how the initial distribution of concentration/temperature decays over time due to diffusion or heat conduction.

$
\text{The sine term in the equation}: \quad \quad  \sin\left(\frac{n\pi x}{l}\right)
$

is the spatial part of the solution, showing how the solution varies along the length $ l $ of the domain.

The summation over $ n $ indicates that this is a series solution, with each term in the series representing a mode of the temperature or concentration distribution in the system.

The $ c_n $ are the Fourier coefficients, which stands for the contribution of each sine wave mode to the initial distribution. Equation 28 provides a formula for calculating these coefficients based on the initial conditions, $ f(x) $, and the steady-state values $ \bar{C}_0 $ and $ \bar{C}_l $.
